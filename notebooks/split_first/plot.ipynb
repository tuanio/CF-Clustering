```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../../data/rfm_train.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 4), ncols=3, nrows=1)
sns.histplot(df, x='Recency', bins=30, ax=ax[0])
sns.histplot(df, x='Frequency', bins=50, ax=ax[1])
sns.histplot(df, x='Moneytary', bins=30, ax=ax[2])
plt.show()
```

```python
cluster_kmeans = pd.read_csv('../data/cluster_kmeans.csv')
cluster_kmeans
```

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(projection='3d')

label = cluster_kmeans['Cluster Label']
for cluster_label in sorted(label.unique()):
    cluster_data = cluster_kmeans[label == cluster_label]
    ax.scatter(cluster_data.Recency,
               cluster_data.Frequency,
               cluster_data.Moneytary,
               label=f'Cluster {cluster_label}')

ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Moneytary')

plt.legend(loc='best')
plt.savefig('3d.jpg')
plt.show()
```
